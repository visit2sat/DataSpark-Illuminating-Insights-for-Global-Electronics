```python
# Remove non-numeric characters and convert to numeric
df3['Zip Code'] = df3['Zip Code'].str.extract('(\d+)', expand=False)

# Convert to numeric, coercing errors to NaN
df3['Zip Code'] = pd.to_numeric(df3['Zip Code'], errors='coerce')

# Fill NaN values (if any) with a default value or handle them as needed
df3['Zip Code'] = df3['Zip Code'].fillna(0).astype(int)


print(df3)
```

```python
q1=df3.quantile(0.25)
q3=df3.quantile(0.50)
iqr=q3-q1
iqr
```

#CHECKING OUTLIERS

```python
def outlier():
    lb=q1-1.5*(iqr)
    ub=q3+1.5*(iqr)
    print(lb)
    print(ub)
```

```python
# prompt: write the code to create table and insert the final2 values

import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user=" root",
  password="",
  database="global_electronics",
  port=3306
)

mycursor = mydb.cursor()

# Create table
mycursor.execute("""
CREATE TABLE IF NOT EXISTS Customers (
  CustomerKey INT,
  Zip_Code INT,
  Gender VARCHAR(255),
  Name VARCHAR(255),
  City VARCHAR(255),
  State VARCHAR(255),
  Country VARCHAR(255),
  Continent VARCHAR(255),
  Birthday DATE
);
""")

# Prepare SQL query to insert data
sql = "INSERT INTO Customers (CustomerKey, Zip_Code, Gender, Name, City, State, Country, Continent, Birthday) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"

# Iterate through DataFrame rows and insert data into the table
for index, row in final2.iterrows():
    values = (
        row['CustomerKey'],
        row['Zip Code'],
        row['Gender'],
        row['Name'],
        row['City'],
        row['State'],
        row['Country'],
        row['Continent'],
        row['Birthday']
    )
    mycursor.execute(sql, values)

mydb.commit()

print(mycursor.rowcount, "record(s) inserted.")
```
